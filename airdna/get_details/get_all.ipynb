```python
import pandas as pd
import json
```

```python

def safe_get(data, keys, default=None):
    """Safely get nested dictionary values"""
    current = data
    for key in keys:
        if isinstance(current, dict) and key in current:
            current = current[key]
        else:
            return default
    return current

# Read JSON from file
with open('../sources/localidades.json', 'r') as file:
    data = json.load(file)

# Extract submarkets data
submarkets = safe_get(data, ['payload', 'submarkets'], [])

# Create DataFrame
records = []
for submarket in submarkets:
    records.append({
        'id': safe_get(submarket, ['id']),
        'name': safe_get(submarket, ['name']),
        'listing_count': safe_get(submarket, ['metrics', 'listing_count'])
    })

df = pd.DataFrame(records)
df.to_csv('../sources/localidades.csv', index=False)
```
